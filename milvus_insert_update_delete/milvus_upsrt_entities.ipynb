# 엔티티 삽입
- 업서트 작업은 데이터 업데이트와 삽입 작업을 결합
- Milvus는 기본 키가 존재하는지 확인하여 업데이트 작업을 수행할지 삽입 작업을 수행할지 결정함

- 참고자료: https://milvus.io/docs/ko/v2.5.x/upsert-entities.md

## 개요
- 컬렉션의 엔티티를 업데이트해야 하거나 업데이트할지 삽입할지 확실하지 않은 경우 Upsert 작업을 사용해 볼 수 있음
- 이 작업을 사용할 때는 Upsert 요청에 포함된 엔티티에 기본 키가 포함되어 있는지 확인해야 하며, 그렇지 않으면 오류가 발생
- Upsert 요청을 받으면 Milvus는 다음 프로세스를 실행
    1. 컬렉션의 기본 필드에 AutoId가 활성화되어 있는지 확인
        1. 활성화된 경우, Milvus는 엔티티의 기본 키를 자동으로 생성된 기본 키로 대체하고 데이터를 삽입
        2. 그렇지 않은 경우 Milvus는 엔티티가 가지고 있는 기본 키를 사용하여 데이터를 삽입
    2. 업서트 요청에 포함된 엔티티의 기본 키 값을 기반으로 삭제 작업을 수행
